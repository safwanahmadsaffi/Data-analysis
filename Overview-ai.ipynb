```python
class student:
#Constructor
  def __init__(self):
    print("Student object created")
    self.id = None
    self.name = None
    self.fee = None
#Destructor

  def __del__(self):
    print("Student object destroyed")

  def user_data(self):
    self.id = input("Enter your ID: ")
    self.name = input("Enter your name: ")
    self.fee = int(input("Enter your you fee: "))

  def display(self):
    print("Student name", self.name)
    print("Student ID", self.id)
    print("Student fee", self.fee)

class course(student):
  university = "Punjab University"

  def __init__(self):
    super().__init__()
    self.course_name = None

  def course_data(self):
    self.course_name = input("Enter your course name: ")

  def display(self):
    super().display()
    print("Course Name: ",self.course_name)
    print("University: ", self.university)

class Hostal(student):

  def __init__(self):
    super().__init__()
    self.room_number = None

  def hostal_data(self):
    self.room_number = input("Enter your room number: ")

  def display(self):
    super().display()
    print("Hostal room number:", self.room_number)


student1 = course()
student1.user_data()
student1.course_data()
student1.display()

student2 = Hostal()
student2.user_data()
student2.hostal_data()
student2.display()
```

```python
class outerclass:
  def __init__(self, outer_value):
    self.outer_value = outer_value

  def display_outer(self):
    print(f"Outer value: {self.outer_value}")

  class InnerClass:
    def __init__(self, inner_value):
      self.inner_value = inner_value

    def display_inner(self):
      print(f"Inner value: {self.inner_value}")

outer = outerclass("Outer")
inner = outerclass.InnerClass("Inner")

outer.display_outer()
inner.display_inner()
```

###Association, Aggregation and Composition

```python
class Address:
  def __init__(self, street, city):
    self.street = street
    self.city = city

class Person:
  def __init__(self, name, address):
    self.name = name
    self.address = address

address = Address("123 abc", "Lahore")
person = Person("Ali", address)

print(f"{person.name}, lives at {person.address.street}, {person.address.city}")
```

```python
#Aggregation
class Book:
  def __init__(self,title):
    self.title = title

class Library:
  def __init__(self):
    self.book = []

  def add_book(self, book):
    self.book.append(book)

book1 = Book("Python")
book2 = Book("Artificial Intelligence")

library = Library()
library.add_book(book1)
library.add_book(book2)

for book in library.book:
  print(book.title)
```

###Polymorphism and Operator Overloading

```python
class Animal:
  def speak(self):
    print("Animal Speaks")

class cat(Animal):
  def speak(self):
    print("Cat meow")

class Dog(Animal):
  def speak(self):
    print("Dogs barks")

def make_animal_speak(animal):
  animal.speak()

cat = cat()
dog = Dog()

make_animal_speak(cat)
make_animal_speak(dog)
```

```python
#Operator Overloading

# +  __add__
# _ __sub__

# / __truediv__

class vector:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return vector (self.x + other.x, self.y + other.y)

  def __repr__(self):
    return f"vector({self.x}, {self.y})"

vector1 = vector(2,3)
vector2 = vector(4,5)

vector3 = vector1 + vector2

print(vector3)
```
